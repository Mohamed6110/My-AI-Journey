#### Exercise 5: Compare Linear vs Polynomial Regression

| Aspect                     | Linear Regression                                   | Polynomial Regression                                      |
|---------------------------|------------------------------------------------------|-------------------------------------------------------------|
| Relationship Type         | Models a straight-line relationship (y = mx + b)     | Models curved/non-linear relationships (y = ax² + bx + c…) |
| Complexity                | Low (simple model)                                   | Higher (more parameters as polynomial degree increases)     |
| When to Use               | Data shows linear trend                               | Data shows nonlinear patterns                               |
| Risk                      | Underfitting if data is nonlinear                     | Overfitting if degree is too high                           |
| Interpretability          | Easy to interpret                                      | Harder to interpret                                         |
| Computation               | Fast, low resources                                   | Slower, higher computational cost                            |


##### Exercise 6: Why Logistic Function Fits Binary Classification

The **logistic (sigmoid) function** fits binary classification because:

- It outputs values between **0 and 1**, which can be interpreted as probabilities.
- It is **smooth and differentiable**, allowing gradient-based optimization.
- It models the probability that a sample belongs to class 1:  
  **P(y=1 | x) = 1 / (1 + e^(-wx))**
- Decision boundary is created at **0.5**, making classification straightforward.
- It prevents outputs from going below 0 or above 1, unlike linear regression.

Therefore, logistic regression is ideal for problems like spam vs. non-spam, fraud vs. non-fraud, etc.

### Exercise 8: Explain MSE, RMSE, MAE

| Metric | Full Name                         | Formula                                      | Meaning / Interpretation                                       | Sensitivity |
|--------|-------------------------------------|----------------------------------------------|------------------------------------------------------------------|-------------|
| MSE    | Mean Squared Error                  | MSE = (1/n) Σ (yᵢ − ŷᵢ)²                      | Penalizes larger errors more because errors are squared.        | Very sensitive to outliers |
| RMSE   | Root Mean Squared Error             | RMSE = √(MSE)                                | Square root of MSE; brings error back to original units.        | Very sensitive to outliers |
| MAE    | Mean Absolute Error                 | MAE = (1/n) Σ |yᵢ − ŷᵢ|                       | Measures average absolute difference between prediction and true value. | Less sensitive to outliers |


### Exercise 9: Compute Accuracy, Precision, Recall, and F1

Given:
- TP = True Positive  
- TN = True Negative  
- FP = False Positive  
- FN = False Negative  

| Metric     | Formula                                                                                  | Meaning |
|------------|-------------------------------------------------------------------------------------------|---------|
| Accuracy   | (TP + TN) / (TP + TN + FP + FN)                                                           | Overall correctness |
| Precision  | TP / (TP + FP)                                                                            | Of all predicted positive samples, how many were correct |
| Recall     | TP / (TP + FN)                                                                            | Of all actual positive samples, how many were detected |
| F1 Score   | 2 × (Precision × Recall) / (Precision + Recall)                                           | Harmonic mean of precision and recall |

### Exercise 12: PCA

**Definition:**  
PCA is a dimensionality reduction technique that transforms high-dimensional data into a lower-dimensional space while preserving as much variance as possible.

**How it works:**  
1. Standardize the data.  
2. Compute the covariance matrix.  
3. Calculate eigenvectors and eigenvalues of the covariance matrix.  
4. Sort eigenvectors by eigenvalues (largest first) to identify principal components.  
5. Project data onto top principal components.

**Goals / Benefits:**  
- Reduce dimensionality to simplify analysis.  
- Remove redundant or correlated features.  
- Improve computational efficiency.  
- Facilitate data visualization (2D, 3D plots).

**Example:**  
- Reducing 50 features of gene expression data to 2–3 principal components for visualization.


### Exercise 13: Hard vs Soft Clustering

| Aspect               | Hard Clustering                                | Soft Clustering                              |
|----------------------|-----------------------------------------------|---------------------------------------------|
| Membership           | Each data point belongs to exactly **one** cluster | Each data point can belong to **multiple clusters** with probabilities |
| Example Algorithms   | K-Means, Hierarchical Clustering             | Fuzzy C-Means, Gaussian Mixture Models (GMM) |
| Output               | Cluster label for each point                   | Membership probabilities for each cluster   |
| Interpretation       | Clear-cut clusters                              | Uncertainty / soft assignment captured      |
| Use Case             | Well-separated, non-overlapping clusters      | Overlapping or ambiguous clusters           |


### Exercise 14: Advantages of DBSCAN

1. **No need to specify number of clusters** in advance (K-Means requires K).  
2. **Can find arbitrarily shaped clusters** (K-Means assumes spherical clusters).  
3. **Handles noise and outliers naturally** (labels them as noise).  
4. **Clusters of different sizes and densities** can be detected.  
5. Useful for datasets with non-globular cluster structure (e.g., spatial data, geolocation).  

**Note:** K-Means may fail when clusters are not well-separated or have irregular shapes.

```python
pca=PCA(n_components=2)
pca.fit(X)
X=pca.transform(X)
```

```python
from sklearn.preprocessing import StandardScaler
X = StandardScaler().fit_transform(X)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Assume 'X_pca' is your 2D PCA-reduced data
# X_pca = ...

max_k = 10
inertia = []
silhouette_scores = []
k_range = range(2, max_k + 1)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    
    # Calculate Silhouette Score (only possible for k > 1)
    if k > 1:
        score = silhouette_score(X, kmeans.labels_)
        silhouette_scores.append(score)

# Plotting the Elbow Method
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method (Inertia)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
# Identify the "elbow" point where the decrease slows down 
# Plotting the Silhouette Scores
plt.subplot(1, 2, 2)
plt.plot(range(2, max_k + 1), silhouette_scores, marker='o')
plt.title('Silhouette Score')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
# Identify the K with the highest score 
plt.show()

# After analysis, set best_k = [Your optimal K]
best_k = 4 # Example, based on the second image
```

```python
from sklearn.metrics import silhouette_score
with mlflow.start_run():
    kmeans = KMeans(n_clusters=best_k,init="k-means++", random_state=42)
    y_kmeans=kmeans.fit_predict(X)
    score = silhouette_score(X, y_kmeans)
    mlflow.log_param("n_clusters", 4)
    mlflow.log_metric("silhouette_score", score)
    mlflow.log_metric("inertia", kmeans.inertia_)
    plt.figure(figsize=(10,10))
    plt.scatter(X[y_kmeans==0,0],X[y_kmeans==0,1],s=50,c='r')
    plt.scatter(X[y_kmeans==1,0],X[y_kmeans==1,1],s=50,c="g")
    plt.scatter(X[y_kmeans==2,0],X[y_kmeans==2,1],s=50,c="b")
    plt.scatter(X[y_kmeans==3,0],X[y_kmeans==3,1],s=50,c='y')
    
    
    plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c="m")
    plt.title("Clusters")

    plt.show()
    plot_path = "kmeans_clusters.png"
    plt.savefig(plot_path)
    mlflow.log_artifact(plot_path, artifact_path="plots")
```

```python
from sklearn.mixture import GaussianMixture

with mlflow.start_run():
    best_k =4
    # Initialize the GMM model
    gmm = GaussianMixture(
        n_components=best_k, 
        covariance_type='full', 
        n_init=10, 
        max_iter=500,
        random_state=42
    )
    # Fit the model and predict the cluster labels
    gmm.fit(X)
    gmm_labels = gmm.predict(X)

    # Plot the GMM results
    plt.figure(figsize=(10, 10))
    plt.scatter(X[:, 0], X[:, 1], c=gmm_labels, cmap='viridis', s=20) 
    plt.title(f'GMM Clustering Results (K={best_k})')
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.show()
```
